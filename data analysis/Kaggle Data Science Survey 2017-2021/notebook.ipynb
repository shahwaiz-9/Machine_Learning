3. FILLING MISSING VALUES  (LEFT -> RIGHT)

```python
# GENERIC FUNCTION

def nullReplacer(columnName):
  null_mask = data[columnName].isnull()
  probs = data[columnName].value_counts(normalize=True)
  data.loc[null_mask, columnName] = np.random.choice(
       probs.index,
       size=null_mask.sum(),
       p=probs.values
  )
```

```python
data['Year'].isnull().sum()
```

```python
data['Total Time'].isnull().sum()
```

```python
# REPLACING NULL WITH AVERAGE TIME
data['Total Time'] = pd.to_numeric(data['Total Time'], errors='coerce')
data['Total Time'] = data['Total Time'].fillna(data['Total Time'].mean())
```

2. Multi Columns analysis and grouping

*WE WILL SPLIT INTO MULTIPLE GROUPS FOR BETTER ANALYSIS*






```python


def analyze_category(df, column_list, title_name):
    """
    Creates a side-by-side analysis dashboard for a group of tools.
    """

    counts = df[column_list].sum().sort_values(ascending=True).reset_index()
    counts.columns = ['Tool', 'Total Users']


    yearly = df.groupby('Year')[column_list].sum().reset_index()
    total_resp = df.groupby('Year').size()
    df_melted = yearly.melt(id_vars='Year', var_name='Tool', value_name='Count')
    df_melted['Market_Share'] = (df_melted['Count'] / df_melted['Year'].map(total_resp)) * 100


    fig = make_subplots(
        rows=1, cols=2,
        subplot_titles=(f"Popularity (Total)", f"Adoption Trends (%)"),
        horizontal_spacing=0.15,
        column_widths=[0.4, 0.6]
    )


    fig.add_trace(
        go.Bar(
            x=counts['Total Users'],
            y=counts['Tool'],
            orientation='h',
            marker=dict(color=counts['Total Users'], colorscale='Viridis'),
            showlegend=False
        ),
        row=1, col=1
    )


    for tool in column_list:
        tool_data = df_melted[df_melted['Tool'] == tool]
        fig.add_trace(
            go.Scatter(
                x=tool_data['Year'],
                y=tool_data['Market_Share'],
                name=tool.split('_')[-1],
                mode='lines+markers',
                line=dict(width=2)
            ),
            row=1, col=2
        )

    fig.update_layout(
        title_text=f"<b>{title_name}</b>",
        height=500,
        template='plotly_white',
        margin=dict(l=50, r=50, t=80, b=50),
        legend=dict(
            orientation="v",
            yanchor="top",
            y=1,
            xanchor="left",
            x=1.02,
            font=dict(size=10)
        )
    )


    display(Markdown(f"---"))
    fig.show()
```

**GROUP 5:** THE NEXT FRONTIER (FUTURE INTENT).

**Objective:** Forecast industry trends.

*These questions (B-Series) identify which technologies respondents plan to adopt or learn in the next 2 years*.

```python


def analyze_future_gap(df, current_cols, future_cols, labels, title_name):
    """
    Compares Current Usage (A-series) vs Future Intent (B-series).
    """
    # 1. Calculate Percentages
    # We use .mean() * 100 because the data is encoded as 0 and 1
    current_vals = df[current_cols].mean() * 100
    future_vals = df[future_cols].mean() * 100

    # 2. Prepare DataFrame for Plotly
    comp_df = pd.DataFrame({
        'Tool': labels,
        'Current Usage (%)': current_vals.values,
        'Future Intent (%)': future_vals.values
    })

    # Calculate Momentum (Gap)
    comp_df['Momentum'] = comp_df['Future Intent (%)'] - comp_df['Current Usage (%)']
    comp_df = comp_df.sort_values(by='Current Usage (%)', ascending=True)

    # 3. Create Subplots
    fig = make_subplots(
        rows=1, cols=2,
        subplot_titles=("Current vs. Future Intent", "Interest Momentum (Gap)"),
        horizontal_spacing=0.15
    )

    # Left: Grouped Bar Chart
    fig.add_trace(go.Bar(y=comp_df['Tool'], x=comp_df['Current Usage (%)'], name='Current (A)', orientation='h', marker_color='#636EFA'), row=1, col=1)
    fig.add_trace(go.Bar(y=comp_df['Tool'], x=comp_df['Future Intent (%)'], name='Future (B)', orientation='h', marker_color='#EF553B'), row=1, col=1)

    # Right: Momentum Chart (Waterfall or Bar)
    colors = ['#00CC96' if x > 0 else '#AB63FA' for x in comp_df['Momentum']]
    fig.add_trace(go.Bar(y=comp_df['Tool'], x=comp_df['Momentum'], orientation='h', marker_color=colors, name='Momentum'), row=1, col=2)

    # 4. Styling
    fig.update_layout(
        title=f"<b>{title_name}: Market Reality vs. Hype</b>",
        height=500, barmode='group', template='plotly_white',
        showlegend=True
    )

    display(Markdown(f"---"))
    fig.show()
```
