```python
# 1. Clean the geolocation table (one lat/long per zip code)
geo_clean = geolocation.groupby('geolocation_zip_code_prefix').agg({
    'geolocation_lat': 'mean',
    'geolocation_lng': 'mean'
}).reset_index()

# 2. Get Customer Coordinates
data = data.merge(
    geo_clean,
    left_on='customer_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
).rename(columns={'geolocation_lat': 'customer_lat', 'geolocation_lng': 'customer_lng'})

# 3. Get Seller Coordinates
data = data.merge(
    geo_clean,
    left_on='seller_zip_code_prefix',
    right_on='geolocation_zip_code_prefix',
    how='left'
).rename(columns={'geolocation_lat': 'seller_lat', 'geolocation_lng': 'seller_lng'})
```

#**PHASE 2 ( Feature Engineering )**

```python
data.shape
```

```python
AnalysisData = data.copy()
```

*Droping unnessecarry columns*

```python
AnalysisData = AnalysisData.drop(columns=["review_comment_title", "review_comment_message", "product_description_length", "product_name_length", ])
```

```python
AnalysisData.shape
```

Introducing new features

```python
# Volume Column as in Brazil volume is used

AnalysisData['Volume_cm3'] = AnalysisData['product_length_cm'] * AnalysisData['product_height_cm'] * AnalysisData['product_width_cm']
```

```python
# Weight Scale

def assign_weight_scale(volume_cm3):
    # Standard Brazilian Freight Factor (6000)
    cubed_weight = volume_cm3 / 6000

    if cubed_weight <= 1:
        return 'Light'
    elif cubed_weight <= 5:
        return 'Medium'
    elif cubed_weight <= 15:
        return 'Heavy'
    else:
        return 'Very Heavy'


AnalysisData['weight_scale'] = AnalysisData['Volume_cm3'].apply(assign_weight_scale)
```

```python
# Actual days seller takes to dispatch order

dispatch_time_days = (AnalysisData['order_delivered_carrier_date'] - AnalysisData['order_approved_at']).dt
dispatch_time_days = dispatch_time_days.days
AnalysisData['dispatch_time_days'] = dispatch_time_days
```

```python
# Actual Carrier time consumed date

carrier_time_days = (AnalysisData['order_delivered_customer_date'] - AnalysisData['order_delivered_carrier_date']).dt
carrier_time_days = carrier_time_days.days
AnalysisData['carrier_time_days'] = carrier_time_days
```

```python
# Actual delivery time consumed

delivery_time_days = (AnalysisData['order_delivered_customer_date'] - AnalysisData['order_purchase_timestamp']).dt
delivery_time_days = delivery_time_days.days
AnalysisData['delivery_time_days'] = delivery_time_days
```

```python
# Estimated delivery time

order_estimated_delivery_days = (AnalysisData['order_estimated_delivery_date'] - AnalysisData['order_purchase_timestamp']).dt
order_estimated_delivery_days = order_estimated_delivery_days.days
AnalysisData['order_estimated_delivery_days'] = order_estimated_delivery_days
```

```python
# Boolean column to check weather delivery was dalayed

order_delayed = AnalysisData['delivery_time_days'] > AnalysisData['order_estimated_delivery_days'].astype(int)
AnalysisData['order_delayed'] = order_delayed
```

```python
# Boolean Column to check seller was delayed

seller_delay = AnalysisData['order_delivered_carrier_date'] > AnalysisData['shipping_limit_date']
AnalysisData['seller_delay'] = seller_delay



# converting to int ( 0/1)
AnalysisData['seller_delay'] = AnalysisData['seller_delay'].astype(int)
AnalysisData['seller_delay']
```

```python
# Distance in Km

def haversine_distance(lat1, lon1, lat2, lon2):
    # Radius of Earth in kilometers
    r = 6371

    # Convert degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat/2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon/2)**2
    c = 2 * np.arcsin(np.sqrt(a))

    return r * c



AnalysisData['distance_km'] = haversine_distance(
    AnalysisData['customer_lat'],
    AnalysisData['customer_lng'],
    AnalysisData['seller_lat'],
    AnalysisData['seller_lng']
)
```

```python
# Distance Scale

def assign_distance_scale(dist):
    if dist <= 50:
        return 'Short'
    elif dist <= 500:
        return 'Medium'
    elif dist <= 1500:
        return 'Large'
    else:
        return 'Very Large'

AnalysisData['distance_scale'] = AnalysisData['distance_km'].apply(assign_distance_scale)
```

```python
# Total Expense

AnalysisData['total_value'] = AnalysisData['payment_value'] + AnalysisData['freight_value']
```

#**PHASE 3 ( Selecting features for Analysis )**

```python
# SINCE WE USED THE JOINS TO MAKE LARGE DATASET THERE WILL BE MULTIPLE DUPLICATES FOR EACH COLUMN
# THE IDEA IS TO EXTRACT COLUMNS WHICH ARE SENSITIVE FOR CERTAIN TASKS
# For example while calucalting total orders we cannot relay on duplicates order id

order_level_bi = AnalysisData.groupby('order_id').agg({
    # --- Existing Columns ---
    'payment_value': 'sum',
    'total_value': 'first',
    'customer_unique_id': 'first',
    'customer_city': 'first',
    'customer_state': 'first',
    'delivery_time_days': 'first',
    'order_estimated_delivery_days': 'first',
    'order_delayed': 'max',
    'review_score': 'mean',
    # --- All Date/Timestamp Columns ---
    'order_purchase_timestamp': 'first',
    'order_approved_at': 'first',
    'order_delivered_carrier_date': 'first',
    'order_delivered_customer_date': 'first',
    'order_estimated_delivery_date': 'first',
    'shipping_limit_date': 'first',
    'review_creation_date': 'first',
    'review_answer_timestamp': 'first'
}).reset_index()
```

```python
item_level_bi = AnalysisData[[
    'order_id', 'product_id', 'seller_id',
    'product_category_name', 'price', 'freight_value',
    'payment_type',
    # --- Physical Attributes ---
    'product_weight_g', 'Volume_cm3', 'weight_scale',
    # --- Logistics & Delay Features ---
    'distance_km', 'distance_scale', 'seller_delay',
    'dispatch_time_days', 'carrier_time_days',
    'order_delivered_customer_date', "order_delivered_carrier_date",
    # --- Seller Info ---
    'seller_city', 'seller_state'
]].copy()
```
